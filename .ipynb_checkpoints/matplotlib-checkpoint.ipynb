## Figure, axes e afins

```python
# Figure é nossa "moldura" do gráfico
plt.figure(figsize=(6, 10), facecolor='blue') #(figsize=[largura, altura])
plt.plot([10, 20, 30, 40], [10, 50, 70, 20], 'go-')
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_axes([0.1, .1, 1, 0.4])

import numpy as np
mu, sigma = 60, 10
x = mu + sigma * np.random.randn(500)

ax.hist(x, bins=50, facecolor='orange', alpha=0.3)
ax.set_xlabel('scores')
ax.set_ylabel('n')
ax.set_title('Histogram', fontsize=18)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.grid(True)
```

**Figure seria como um mural, e o add_axes seria o espaço da foto.**

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_axes([0.1, .1, 1, 0.4])

import numpy as np
mu, sigma = 60, 10
x = mu + sigma * np.random.randn(500)

ax.hist(x[x > 50], bins=30, facecolor='orange', alpha=0.8);
ax.hist(x[x <= 50], bins=30, facecolor='orange', alpha=0.4);
# ; tira os vetores com informações
```

## Outros gráficos

```python
plt.figure(figsize=(8,4))
plt.bar(x=[1, 2, 3, 4],
        height=[1, 4, 9, 16],
        color='r',
        width=0.8,
        edgecolor='orange',
        tick_label=['Barra A', 'Barra B', 'Barra C', 'Barra D']
        # bottom=5
        # log=True
);

plt.xlabel("Eixo x", fontsize=12)
plt.xlabel("Eixo y", fontsize=12)
plt.title("Grafico de barras", fontsize=18)
```

```python
plt.barh([1, 2, 3, 4], [1, 2, 3, 4])
```

```python
graf_barra = plt.bar([1, 2, 3, 4], [1, 2, 3, 4], tick_label=['Grupo a', 'Grupo B', 'Grupo C', 'Grupo D'])
graf_barra[0].set_color('red')
graf_barra[1].set_color('orange')
graf_barra[2].set_color('purple')
```
