```python
import seaborn as sns
import pandas as pd
```

```python
iris = pd.read_csv('/content/Iris.csv')
iris.head(5)
```

```python
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=iris)
```

```python
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, hue='Species')
```

```python
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, hue='Species', style='Species')
```

```python
sns.barplot(x='PetalLengthCm', y='Species', data=iris, errorbar=('ci', 20));
```

```python
sns.barplot(y='PetalLengthCm', x='Species', data=iris, errorbar=('ci', 20));
```

```python
import numpy as np
iris['LocalColheita'] = np.where(iris['Id']%2==0, 'SP', 'MG')
sns.barplot(x='Species', y='PetalLengthCm', hue='LocalColheita', data=iris, errorbar=('ci', 20));
```

```python
sns.barplot(y='PetalLengthCm', x='Species', data=iris, errorbar=('ci', 20), palette='Blues_d');
```
